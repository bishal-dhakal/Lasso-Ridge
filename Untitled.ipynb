```python
##R_squared
def adj_r2(X,y):
    r2=regression.score(X,y)
    n=X.shape[0]
    p=X.shape[1]
    adjusted_r2=1-(1-r2)*(n-1)/(n-p-1)
    return adjusted_r2
```
